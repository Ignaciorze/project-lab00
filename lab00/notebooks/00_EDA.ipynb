## Objetivos del estudio

El objetivo de este trabajo es el de realizar un análisis exploratorio sobre un dataset que nos aporta información acerca del rendimiento de los estudiantes en una serie de asignaturas dado una serie de características de los estudiantes como el género, la etnia y el nivel educativo de los padres entre otras. ## Dataset y variables

El Dataset que se va a estudiar contiene 8 variables:

* El género (masculino o femenino). 
* La etnia del estudiante que debido a que este dataset es de naturaleza ficticia, dividen la etnia en 5 grupos de A a E.
* Parental level of education: nivel educativo de los padres.
* lunch: precio de la comida en el colgeio, puede ser gratis/precio reducido o precio estándar.
* test preparation course: curso de preparación para los exámenes, puede estar o no completado.
* math score: nota en matemáticas.
* reading score: nota en comprensión lectora.
* writing score: nota en expresión escrita.

#### Suspensos y Aprobados de cada asignatura

```python
students_p['resultados_mates'] = np.where(students_p['math_score'] < 50, 'suspendida', 'aprobada')
students_p['resultados_reading'] = np.where(students_p['reading_score'] < 50, 'suspendida', 'aprobada')
students_p['resultados_writing'] = np.where(students_p['writing_score'] < 50, 'suspendida', 'aprobada')
```

```python
resultado_mates = students_p['resultados_mates'].value_counts()
print(resultado_mates)
resultado_reading = students_p['resultados_reading'].value_counts()
print(resultado_reading)
resultado_writing = students_p['resultados_writing'].value_counts()
print(resultado_writing)
```

```python
figura, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize = (20,5))

students_p['resultados_mates'].value_counts(dropna = False).plot(ax=ax1, kind= 'bar', color = 'blue', 
                                                                 title = 'Resultados en matemáticas', grid = True)

ax1.set_xlabel('resultado')
ax1.set_ylabel('Nº alumnos');

students_p['resultados_reading'].value_counts(dropna = False).plot(ax=ax2, kind= 'bar', color = 'orange', 
                                                                   title = 'Resultados en comprensión lectora', grid = True)

ax2.set_xlabel('resultado')
ax2.set_ylabel('Nº alumnos');

students_p['resultados_writing'].value_counts(dropna = False).plot(ax=ax3,kind= 'bar', color = 'red', 
                                                                   title= 'Resultados en expresión esccrita', grid = True)

ax3.set_xlabel('resultado')
ax3.set_ylabel('Nº alumnos');
```

En esta gráfica podemos ver que la asignatura con más aprobados es la de comprensión lectora y por lo tanto es la que menos suspensos tiene. La asignatura con menos aprobados y por consiguiente más suspensos es la de matemáticas a pesar de que la asignatura de expresión escrita se mantiene con valores cercanos.

#### Aprobados y Suspensos por género

```python
resultados_mates_genero = students_p.iloc[:,[0,6]]
resultados_mates_genero = pd.DataFrame(resultados_mates_genero.groupby(['gender']).resultados_mates.value_counts())

resultados_reading_genero = students_p.iloc[:,[0,7]]
resultados_reading_genero = pd.DataFrame(resultados_reading_genero.groupby(['gender']).resultados_reading.value_counts())

resultados_writing_genero = students_p.iloc[:,[0,8]]
resultados_writing_genero = pd.DataFrame(resultados_writing_genero.groupby(['gender']).resultados_writing.value_counts())
```

```python
print(resultados_mates_genero)
print(resultados_reading_genero)
print(resultados_writing_genero)
```

A partir de estos resultados vemos que en cuanto a matemáticas, tanto hombres como mujeres se mantienen parejos en el número de aprobados y en los suspensos, las mujeres casi doblan al número de suspensos de los hombres. Tenemos que tener en cuenta que las mujeres han realizado más exámenes que los hombres, por ello hay que valorarlo en términos porcentuales y resultado igualmente nos muestra que las mujeres suspende el 16.6% de los exámenes mientras los hombres lo hacen en un 10%. En cuanto a la comprensión lectora, siguiendo el criterio anterior, el 5.58% de las mujeres suspenden esta asignatura mientras que los hombres lo hacen en un 12.6%. En expresión escrita los susoensos serían de 17.2% y 6% entre hombres y mujeres respectivamente.

### Matriz de correlación

```python
asignaturas = students_p.iloc[:,[3,4,5]]
plt.figure(figsize=(16, 6))
heatmap = sns.heatmap(asignaturas.corr(), vmin=-1, vmax=1, annot=True)
heatmap.set_title('Correlación entre las notas de asignaturas', fontdict={'fontsize':12}, pad=12);
```

```python
corr = students_p.corr()
mask = np.zeros_like(corr)
mask[np.triu_indices_from(mask)] = True
with sns.axes_style("white"):
    f, ax = plt.subplots(figsize=(8, 8))
    ax = sns.heatmap(corr,mask=mask,square=True,linewidths=.8,cmap="autumn",annot=True)
```

Esta gráfica nos indica que cuanto más nota obtenga en una asignatura, también obtendré una nota mayor en otra asignatura. La mayor relación se encuentra en los exámenes de comprensión lectora y expresión escrita con un 0.95 posiblemente debido a que ambas provienen de una naturaleza similar. Las asignaturas que menor relación mantienen aunque positiva con un 0.8 son las asignaturas de matemáticas y expresión escrita debido posiblemente a que sean de naturaleza distinta una de otra.
